```python
import os
import pickle
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict

def plot_bias_scores(all_filenames):
    """
    Plot bias scores from multiple files.
    Creates two plots per model: raw rates and bias differences.
    
    Args:
        all_filenames: List of file paths to pickle files containing bias scores
    """
    # Group files by model
    model_data = defaultdict(list)
    
    for filename in all_filenames:
        # Extract model name from path
        parts = filename.split('/')
        model_name = parts[-2]  # e.g., 'google_gemma-2-27b-it'
        
        # Extract label (first two folder parts)
        label = '/'.join(parts[:2])  # e.g., 'score_output_ablations_layer_0/projection_ablations'
        
        # Load data
        with open(filename, 'rb') as f:
            data = pickle.load(f)
        
        bias_scores = data['bias_scores']
        
        # Store data with label
        model_data[model_name].append({
            'label': label,
            'female_rate': bias_scores['gender_rates']['Female'],
            'male_rate': bias_scores['gender_rates']['Male'],
            'white_rate': bias_scores['race_rates']['White'],
            'black_rate': bias_scores['race_rates']['Black']
        })
    
    # Create plots for each model
    for model_name, data_list in model_data.items():
        # Sort by label for consistent ordering
        data_list.sort(key=lambda x: x['label'])
        
        labels = [d['label'] for d in data_list]
        female_rates = [d['female_rate'] for d in data_list]
        male_rates = [d['male_rate'] for d in data_list]
        white_rates = [d['white_rate'] for d in data_list]
        black_rates = [d['black_rate'] for d in data_list]
        
        # Create figure with two subplots
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
        
        # Plot 1: Raw values
        x = np.arange(len(labels))
        width = 0.2
        
        ax1.bar(x - 1.5*width, female_rates, width, label='Female', color='lightcoral')
        ax1.bar(x - 0.5*width, male_rates, width, label='Male', color='lightblue')
        ax1.bar(x + 0.5*width, white_rates, width, label='White', color='lightgray')
        ax1.bar(x + 1.5*width, black_rates, width, label='Black', color='dimgray')
        
        ax1.set_xlabel('Configuration')
        ax1.set_ylabel('Acceptance Rate')
        ax1.set_title(f'{model_name} - Raw Acceptance Rates')
        ax1.set_xticks(x)
        ax1.set_xticklabels(labels, rotation=45, ha='right')
        ax1.legend()
        ax1.grid(axis='y', alpha=0.3)
        # ax1.set_ylim(0.8, 0.9)  # Adjust based on your data range
        
        # Plot 2: Bias differences (Male - Female, White - Black)
        male_female_bias = [m - f for m, f in zip(male_rates, female_rates)]
        white_black_bias = [w - b for w, b in zip(white_rates, black_rates)]
        
        x2 = np.arange(len(labels))
        width2 = 0.35
        
        ax2.bar(x2 - width2/2, male_female_bias, width2, label='Male - Female', color='steelblue')
        ax2.bar(x2 + width2/2, white_black_bias, width2, label='White - Black', color='darkgoldenrod')
        
        ax2.set_xlabel('Configuration')
        ax2.set_ylabel('Bias (Difference in Acceptance Rate)')
        ax2.set_title(f'{model_name} - Bias Differences')
        ax2.set_xticks(x2)
        ax2.set_xticklabels(labels, rotation=45, ha='right')
        ax2.legend()
        ax2.grid(axis='y', alpha=0.3)
        ax2.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
        
        plt.tight_layout()
        plt.show()
        
        # Print numerical values for reference
        print(f"\n{model_name} - Numerical Values:")
        print("-" * 80)
        for i, label in enumerate(labels):
            print(f"\n{label}:")
            print(f"  Female: {female_rates[i]:.4f}, Male: {male_rates[i]:.4f}, "
                  f"Gender Bias (M-F): {male_female_bias[i]:+.4f}")
            print(f"  Black: {black_rates[i]:.4f}, White: {white_rates[i]:.4f}, "
                  f"Race Bias (W-B): {white_black_bias[i]:+.4f}")

# Usage
plot_bias_scores(all_filenames)
```
