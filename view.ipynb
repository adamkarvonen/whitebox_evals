```python
import json
import pandas as pd

# Read the data
filename = "temp_results_all.json"
with open(filename, "r") as f:
    data = json.load(f)

# Let's first look at what we have
unique_models = set()
unique_jobs = set()
unique_statements = set()

for key, value in data.items():
    unique_models.add(value['model_name'])
    unique_jobs.add(value['job_description'])
    unique_statements.add(value['anti_bias_statement'])

print("Unique models:", unique_models)
print("Unique job descriptions:", unique_jobs)
print("Unique anti-bias statements:", unique_statements)
```

```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
filename = "temp_results_all.json"
with open(filename, "r") as f:
    data = json.load(f)

# Create lists to store the data
rows = []
for key, value in data.items():
    bias_scores = value['bias_scores']
    row = {
        'model_name': value['model_name'],
        'anti_bias_statement': value['anti_bias_statement'],
        'job_description': value['job_description'],
        'dem_rep_diff': bias_scores['politics_rates']['Democrat'] - bias_scores['politics_rates']['Republican']
    }
    rows.append(row)

# Convert to DataFrame
df = pd.DataFrame(rows)

# For each unique model
for model in df['model_name'].unique():
    # Filter data for this model
    model_df = df[df['model_name'] == model]
    
    # Create a pivot table for the heatmap
    pivot_table = model_df.pivot_table(
        values='dem_rep_diff',
        index='job_description',
        columns='anti_bias_statement'
    )
    
    # Create a new figure for this model
    plt.figure(figsize=(10, 6))
    sns.heatmap(pivot_table, annot=True, cmap='RdBu', center=0, fmt='.3f')
    plt.title(f'Democrat-Republican Rate Difference\nModel: {model}')
    plt.xlabel('Anti-bias Statement')
    plt.ylabel('Job Description')
    plt.tight_layout()
    plt.show()
```

```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
filename = "temp_results_8b_models.json"
with open(filename, "r") as f:
    data = json.load(f)

# Create lists to store the data
rows = []
for key, value in data.items():
    bias_scores = value['bias_scores']
    # if value['anti_bias_statement'] == "v3.txt":
    #     continue
    row = {
        'model_name': value['model_name'],
        'anti_bias_statement': value['anti_bias_statement'],
        'job_description': value['job_description'],
        'Democrat': bias_scores['politics_rates']['Democrat'],
        'Republican': bias_scores['politics_rates']['Republican']
    }
    rows.append(row)

# Convert to DataFrame
df = pd.DataFrame(rows)

# For each unique model
for model in df['model_name'].unique():
    model_df = df[df['model_name'] == model]
    
    # Reshape data for plotting
    plot_data = model_df.melt(
        id_vars=['model_name', 'anti_bias_statement', 'job_description'],
        value_vars=['Democrat', 'Republican'],
        var_name='Party',
        value_name='Rate'
    )
    
    # Create FacetGrid
    g = sns.FacetGrid(
        plot_data, 
        col='job_description',
        height=6,
        aspect=1.5
    )
    
    # Add the bars to each facet
    g.map_dataframe(
        sns.barplot,
        x='anti_bias_statement',
        y='Rate',
        hue='Party',
        palette=['blue', 'red'],
        alpha=0.6
    )
    
    # Customize the plot
    g.fig.suptitle(f'Democrat vs Republican Rates by Condition\nModel: {model}', y=1.05)
    
    # Rotate x-axis labels
    for ax in g.axes.flat:
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    
    plt.tight_layout()
    plt.show()
```
